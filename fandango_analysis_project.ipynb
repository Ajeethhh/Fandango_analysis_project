```python
df = pd.merge(fandango,all_sites,how='inner')
```

```python
df
```

```python
df['RT_norm'] = np.round((df['RottenTomatoes']-df['RottenTomatoes'].mean())/df['RottenTomatoes'].std(),1)
```

```python
df
```

```python
df = df.drop('rt_norm',axis=1)
```

```python
df
```

```python
df['RT_norm'] = np.round(df['RottenTomatoes']/20,1)
```

```python
df['RTU_norm'] = np.round(df['RottenTomatoes_User']/20,1)
```

```python
df['MC_norm'] = np.round(df['Metacritic']/20,1)
```

```python
df['MCU_norm'] = np.round(df['Metacritic_User']/2,1)
```

```python
df['IMDB_norm'] = np.round(df['IMDB']/2,1)
```

```python
df
```

```python
norm_scores = df[['FILM','STARS','RATING','RT_norm','RTU_norm','MC_norm','MCU_norm','IMDB_norm']]
```

```python
norm_scores 
```

```python
plt.figure(figsize=(15,6),dpi=150)
sns.kdeplot(data=norm_scores,fill=True,clip=[0,5],palette='Set1')
```

```python
plt.figure(figsize=(15,6),dpi=150)
sns.histplot(data=norm_scores,bins=45)
```

```python
plt.figure(figsize=(1,1))
sns.clustermap(data=norm_scores.drop('FILM',axis=1),col_cluster=False)
```
